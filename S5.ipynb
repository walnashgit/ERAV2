```python
from gdrive.MyDrive.ERA.model import Network
from gdrive.MyDrive.ERA.utils import *


# Setup training data
train_transform = getTrainingTransforms()
train_data = getTrainingData(train_transform)

# Setup test data
test_transform = getTestTransforms()
test_data = getTestData(test_transform)

# Set the batch size for taining and test
batch_size = 512

# Setup training dataLoader
train_loader = getDataLoader(train_data, batch_size)

# Setup test dataLoader
test_loader = getDataLoader(test_data, batch_size)

# Plot the training sample data for visualization, I am plotting 9 data
plotTrainDataSample(train_loader, 9)

# Get the available device
device = getAvailableDevice()

# Create the instance of the model and install in the device
model = Network().to(device)

# Print the summary of the model
printModelSummary(model)

#
# TRAIN and TEST the model
#

# get the optimizer
optimizer = getOptimizer(model)
# get the scheduler
scheduler = getScheduler(optimizer)
# get the criterion
criterion = getCriterion()

num_epochs = 20

for epoch in range(1, num_epochs+1):
  print(f'Epoch {epoch}')
  train(model, device, train_loader, optimizer, criterion)
  test(model, device, train_loader, criterion)
  scheduler.step()

# Plot trainig's and test's loss and accuracy data
plotTrainingTestLossAndAccuracy()

```
