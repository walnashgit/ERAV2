```python
from ERAV2_main.model.S11ResNet_model import ResNet18
from ERAV2_main.utils import CIFAR10ResNetUtil

import torch.optim as optim
import torch.nn as nn

util = CIFAR10ResNetUtil()

train_set = util.get_train_set_cifar10()
train_loader = util.get_data_loader_cifar10(train_set)

test_set = util.get_test_set_cifar10()
test_loader = util.get_data_loader_cifar10(test_set)

model = ResNet18()
device = util.get_available_device()

util.print_summary(model, device)

optimizer = optim.Adam(model.parameters(), lr=1e-10, weight_decay=1e-2)
criterion = nn.CrossEntropyLoss()
lr = util.find_lr_fastai(model, device, train_loader, criterion, optimizer)

lr_max = round(lr, 6)
lr_min = lr_max/10
print("lr_max: ", lr_max)

# scheduler = optim.lr_scheduler.CyclicLR(optimizer, base_lr=lr_min, max_lr=lr_max, step_size_up=4, step_size_down=19,
#                                         mode="triangular", cycle_momentum=False)

EPOCHS = 20

# scheduler = optim.lr_scheduler.OneCycleLR(
#         optimizer,
#         max_lr=lr_max,
#         steps_per_epoch=len(train_loader),
#         epochs=EPOCHS,
#         pct_start=.3,
#         div_factor=100,
#         three_phase=False,
#         final_div_factor=100,
#         anneal_strategy='linear'
#     )

scheduler = optim.lr_scheduler.CyclicLR(optimizer, base_lr=lr_min, max_lr=lr_max, step_size_up=5, step_size_down=14,
                                        mode="triangular", cycle_momentum=False)

collect_images = True
util.start_training_testing(epochs=EPOCHS, collect_images=collect_images, model=model, device=device,
                            train_loader=train_loader,
                            test_loader=test_loader,
                            optimizer=optimizer,
                            criterion=criterion,
                            scheduler=scheduler
                            )

util.plot_miss_classified_images(train_set=train_set)

util.show_grad_cam_heatmap(model=model, train_set=train_set)

util.plot_training_and_test_loss_and_accuracy()

```
